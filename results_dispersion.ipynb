```python
import pandas as pd
import seaborn as sns
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('sim_result_ntrials_0.pickle', 'rb') as f:
    data_0 = pickle.load(f)

with open('sim_result_ntrials_1.pickle', 'rb') as f:
    data_1 = pickle.load(f)
```

```python
df_0 = pd.DataFrame(data_0)
df_1 = pd.DataFrame(data_1)
df_1.head(5)
```

```python
fig, ax = plt.subplots(2, 2, figsize=(16, 6))
sns.histplot(data=df_0, x='process_time_mean', hue='time', ax=ax[0, 0], element="poly", bins=9)
sns.histplot(data=df_0, x='process_time_std', hue='time', ax=ax[0, 1], element="poly", bins=9)
sns.histplot(data=df_0, x='small_load', hue='time', ax=ax[1, 0], element="poly", bins=9)
sns.histplot(data=df_0, x='big_load', hue='time', ax=ax[1, 1], element="poly", bins=9)
plt.suptitle('  ')
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(16, 6))
sns.histplot(data=df_1, x='process_time_mean', hue='time', ax=ax[0, 0], element="poly", bins=9)
sns.histplot(data=df_1, x='process_time_std', hue='time', ax=ax[0, 1], element="poly", bins=9)
sns.histplot(data=df_1, x='small_load', hue='time', ax=ax[1, 0], element="poly", bins=9)
sns.histplot(data=df_1, x='big_load', hue='time', ax=ax[1, 1], element="poly", bins=9)
plt.suptitle('  ')
plt.show()
```

```python
fig, ax = plt.subplots(1, 3, figsize=(18, 4))
sns.histplot(data=df_0.loc[df_0.time == 10000], x='process_time_mean', hue='time', ax=ax[0], bins=5)
sns.histplot(data=df_0.loc[df_0.time == 10000], x='small_load', hue='time', ax=ax[1], bins=5)
sns.histplot(data=df_0.loc[df_0.time == 10000], x='big_load', hue='time', ax=ax[2], bins=5)
plt.suptitle('  ')
plt.show()
```

```python
fig, ax = plt.subplots(1, 3, figsize=(18, 4))
sns.histplot(data=df_1.loc[df_1.time == 10000], x='process_time_mean', hue='time', ax=ax[0], bins=5)
sns.histplot(data=df_1.loc[df_1.time == 10000], x='small_load', hue='time', ax=ax[1], bins=5)
sns.histplot(data=df_1.loc[df_1.time == 10000], x='big_load', hue='time', ax=ax[2], bins=5)
plt.suptitle('  ')
plt.show()
```
