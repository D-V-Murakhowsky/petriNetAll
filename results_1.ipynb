```python
import pickle
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
with open('sim_result_1.pickle', 'rb') as f:
    data = pickle.load(f)
```

```python
append = np.array(data['Place_OnConveyor']['append'][:len(data['Place_OnConveyor']['exclude'])])
exclude = np.array(data['Place_OnConveyor']['exclude'])
wait_time = exclude - append
```

```python
sns.histplot(wait_time, bins=10)
plt.show()
```

```python
small_load = data['Place_SmallFree']['current_load']
big_load = data['Place_BigFree']['current_load']
```

```python
print(f'Середнє завантаження маленьких літаків: {3 - small_load[:, 1].mean():.2f}')
print(f'Середнє завантаження великих літаків: {2 - big_load[:, 1].mean():.2f}')
```
