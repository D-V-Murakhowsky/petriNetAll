```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
with open('left.npy', 'rb') as f:
    left_stats = np.load(f)
with open('productivity.npy', 'rb') as f:
    productivity_stats = np.load(f)
```

```python
left_stats = left_stats.mean(axis=1)
productivity_stats = productivity_stats.mean(axis=1)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(18, 4))
sns.lineplot(x=np.linspace(1000, 50000, 50), y=left_stats, ax=ax[0])
sns.lineplot(x=np.linspace(1000, 50000, 50), y=productivity_stats / 20, ax=ax[1])
ax[0].set_xlabel(' , .')
ax[0].set_ylabel('   , .')
ax[1].set_xlabel(' , .')
ax[1].set_ylabel(' , ./.')
plt.show()
```
