```python
import pandas as pd
import seaborn as sns
import pickle
import matplotlib.pyplot as plt
```

```python
with open('sim_result_data_0.pickle', 'rb') as f:
    data_0 = pickle.load(f)

with open('sim_result_data_1.pickle', 'rb') as f:
    data_1 = pickle.load(f)
```

```python
df_0 = pd.DataFrame(data_0)
df_1 = pd.DataFrame(data_1)
df_1.head(5)
```

```python
fig, ax = plt.subplots(2, 2, figsize=(16, 6))
sns.lineplot(data=df_0, x='time', y='process_time_mean', ax=ax[0, 0])
sns.lineplot(data=df_0, x='time', y='process_time_std', ax=ax[0, 1])
sns.lineplot(data=df_0, x='time', y='small_load', ax=ax[1, 0])
sns.lineplot(data=df_0, x='time', y='big_load', ax=ax[1, 1])
plt.suptitle('Залежності відгуків моделі від часу')
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(16, 6))
sns.lineplot(data=df_1, x='time', y='process_time_mean', ax=ax[0, 0])
sns.lineplot(data=df_1, x='time', y='process_time_std', ax=ax[0, 1])
sns.lineplot(data=df_1, x='time', y='small_load', ax=ax[1, 0])
sns.lineplot(data=df_1, x='time', y='big_load', ax=ax[1, 1])
plt.suptitle('Залежності відгуків моделі від часу')
plt.show()
```
