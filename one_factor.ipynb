```python
import pandas as pd
import seaborn as sns
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('sim_result_ncontainers_0.pickle', 'rb') as f:
    data_0 = pickle.load(f)

with open('sim_result_ncontainers_1.pickle', 'rb') as f:
    data_1 = pickle.load(f)
```

```python
df_0 = pd.DataFrame(data_0)
df_1 = pd.DataFrame(data_1)
df_1.head(5)
```

```python
df_agg = df_0.groupby('n').agg({'process_time_mean': np.mean, 'small_load': np.mean, 'big_load': np.mean}).reset_index(drop=False)
df_agg_1 = df_1.groupby('n').agg({'process_time_mean': np.mean, 'small_load': np.mean, 'big_load': np.mean}).reset_index(drop=False)
df_agg
```

```python
fig, ax = plt.subplots(3, 2, figsize=(14, 12), sharex=True)
sns.lineplot(data=df_agg, x='n', y='process_time_mean', ax=ax[0, 0])
sns.lineplot(data=df_agg_1, x='n', y='process_time_mean', ax=ax[0, 1])
sns.lineplot(data=df_agg, x='n', y='small_load', ax=ax[1, 0])
sns.lineplot(data=df_agg_1, x='n', y='small_load', ax=ax[1, 1])
sns.lineplot(data=df_agg, x='n', y='big_load', ax=ax[2, 0])
sns.lineplot(data=df_agg_1, x='n', y='big_load', ax=ax[2, 1])
plt.show()
```

```python
with open('sim_result_time_0.pickle', 'rb') as f:
    data_0 = pickle.load(f)

with open('sim_result_time_1.pickle', 'rb') as f:
    data_1 = pickle.load(f)

df_0 = pd.DataFrame(data_0)
df_1 = pd.DataFrame(data_1)
df_agg_0 = df_0.groupby('time').agg({'process_time_mean': np.mean}).reset_index()
df_agg_1 = df_1.groupby('time').agg({'process_time_mean': np.mean}).reset_index()
fig, ax = plt.subplots(1, 2, figsize=(14, 4))
sns.lineplot(data=df_agg_0, x='time', y='process_time_mean', ax=ax[0])
sns.lineplot(data=df_agg_1, x='time', y='process_time_mean', ax=ax[1])
plt.show()
```
