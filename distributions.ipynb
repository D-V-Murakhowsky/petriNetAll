```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
with open('left.npy', 'rb') as f:
    left_stats = np.load(f)
with open('productivity.npy', 'rb') as f:
    productivity_stats = np.load(f)
```

```python
left_stats.ravel().shape
```

```python
fig, ax = plt.subplots(1, 2, figsize=(18, 4))
sns.histplot(x=left_stats.ravel(), bins=20, ax=ax[0])
sns.histplot(x=productivity_stats.ravel(), bins=20, ax=ax[1])
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(18, 4))
sns.histplot(x=left_stats.ravel(), bins=10, ax=ax[0])
sns.histplot(x=productivity_stats.ravel(), bins=10, ax=ax[1])
plt.show()
```
